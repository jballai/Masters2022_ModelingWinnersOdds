```python
import pandas as pd
import numpy as np

arg = pd.read_csv("around_grn.csv")
acc = pd.read_csv("driv_acc.csv")
dist = pd.read_csv("driv_dist.csv")
putt = pd.read_csv("sg_putting.csv")
approach = pd.read_csv("sg_approach.csv")

arg = arg.drop(['web-scraper-order', 'web-scraper-start-url'], axis = 1)
acc = acc.drop(['web-scraper-order', 'web-scraper-start-url'], axis = 1)
dist = dist.drop(['web-scraper-order', 'web-scraper-start-url'], axis = 1)
putt = putt.drop(['web-scraper-order', 'web-scraper-start-url'], axis = 1)
approach = approach.drop(['web-scraper-order', 'web-scraper-start-url'], axis = 1)
```

```python
dist = dist.apply(lambda x: x/dist['avg_driv_dist'].sum() if x.name == 'avg_driv_dist' else x)
acc = acc.apply(lambda x: x/acc['pct_fairways'].sum()  if x.name == 'pct_fairways' else x)
approach = approach.apply(lambda x: x/approach['TOTAL SG_APP'].sum() if x.name == 'TOTAL SG_APP' else x)
arg = arg.apply(lambda x: x/arg['sg_arg'].sum() if x.name == 'sg_arg' else x)
putt = putt.apply(lambda x: x/putt['sg_putting'].sum() if x.name == 'sg_putting' else x)
```

```python
w_dist = 0.625
w_acc = 0.25
w_app = 0.97
w_arg = 0.77
w_putt = 0.6

dist = dist.apply(lambda x: w_dist*x if x.name == 'avg_driv_dist' else x)
acc = acc.apply(lambda x: w_acc*x if x.name == 'pct_fairways' else x)
approach = approach.apply(lambda x: w_app*x if x.name == 'TOTAL SG_APP' else x)
arg = arg.apply(lambda x: w_arg*x if x.name == 'sg_arg' else x)
putt = putt.apply(lambda x: w_putt*x if x.name == 'sg_putting' else x)
```

```python

merged = pd.merge(arg, acc)
merged = pd.merge(merged, dist)
merged = pd.merge(merged, putt)
merged = pd.merge(merged, approach)
merged.head()
```

```python
merged['SUM'] = merged['sg_arg'] + merged['pct_fairways'] + merged['avg_driv_dist'] + merged['sg_putting'] + merged['TOTAL SG_APP']
merged.head()
```

```python
merged = merged.sort_values(by=['SUM'], ascending=False)
```

```python
merged = merged.apply(lambda x: x/merged['SUM'].sum() if x.name == 'SUM' else x)
merged

#  Final output of 'SUM' category gives the percent chance the golfer has to win the tournament.
```
